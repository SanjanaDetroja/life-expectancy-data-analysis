# **Step 1: Data Cleaning**

```python
df.info()
```

```python
# checking for no of rows and columns.

print('No of rows in dataframe',df.shape[0])
print('No of columns in dataframe',df.shape[1]) 
```

```python
#checking for null values:

df.isnull().sum()
```

```python
#imputation done on the basics of individual columns with mean and median

df["Life expectancy "] = df["Life expectancy "].replace(np.NaN, df["Life expectancy "].mean())
df["infant deaths"] = df["infant deaths"].replace(np.NaN, df["infant deaths"].median())
df["Adult Mortality"] = df["Adult Mortality"].replace(np.NaN, df["Adult Mortality"].median())
df["Alcohol"] = df["Alcohol"].replace(np.NaN, df["Alcohol"].median())
df["Hepatitis B"] = df["Hepatitis B"].replace(np.NaN, df["Hepatitis B"].median())
df[" BMI "] = df[" BMI "].replace(np.NaN, df[" BMI "].mean())
df["Polio"] = df["Polio"].replace(np.NaN, df["Polio"].mean())
df["Total expenditure"] = df["Total expenditure"].replace(np.NaN, df["Total expenditure"].mean())
df["Diphtheria "] = df["Diphtheria "].replace(np.NaN, df["Diphtheria "].mean())
df["GDP"] = df["GDP"].replace(np.NaN, df["GDP"].median())
df[" thinness  1-19 years"] = df[" thinness  1-19 years"].replace(np.NaN, df[" thinness  1-19 years"].median())
df[" thinness 5-9 years"] = df[" thinness 5-9 years"].replace(np.NaN, df[" thinness 5-9 years"].median())
df["Income composition of resources"] = df["Income composition of resources"].replace(np.NaN, df["Income composition of resources"].median())
df["Schooling"] = df["Schooling"].replace(np.NaN, df["Schooling"].median())
df["Population"] = df["Population"].replace(np.NaN, df["Population"].mean())
```

```python
#checking for null values:

df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a seaborn style for better aesthetics
sns.set(style="whitegrid")

# Insight 2: Immunization Coverage
immunization_columns = ['Hepatitis B', 'Polio', 'Diphtheria ']

# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[immunization_columns], orient='h', palette='viridis')

# Customize the plot
#plt.title('Distribution of Immunization Coverage')
plt.xlabel('Coverage (%)')
plt.ylabel('Vaccines')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a seaborn style for better aesthetics
sns.set(style="whitegrid")

# Insight 4: Relationship between Alcohol Consumption and Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Alcohol', y='Life expectancy ', data=df, color='skyblue', edgecolor='black', alpha=0.7)

# Customize the plot
plt.xlabel('Alcohol Consumption (Liters per capita)')
plt.ylabel('Life Expectancy (Years)')
plt.title('Relationship between Alcohol Consumption and Life Expectancy')

# Add a grid for better readability
plt.grid(linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Set a seaborn style for better aesthetics
sns.set(style="whitegrid")

# Assuming your dataset is in a DataFrame named 'df'
# If your data is in a CSV file, you can read it using: df = pd.read_csv('your_data.csv')

# Select numerical columns for scaling and box plot
numerical_columns = ['Life expectancy ', 'Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure',
                     'Hepatitis B', 'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
                     'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population', ' thinness  1-19 years', ' thinness 5-9 years',
                     'Income composition of resources', 'Schooling']

# Min-Max scaling
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[numerical_columns]), columns=numerical_columns)

# Create box plots for scaled numerical columns
plt.figure(figsize=(16, 10))
sns.boxplot(data=df_scaled, orient='h', palette='pastel')

# Customize the plot
plt.title('Box Plots for Scaled Numerical Columns (Min-Max Scaling)', fontsize=16)
plt.xlabel('Scaled Values', fontsize=12)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
